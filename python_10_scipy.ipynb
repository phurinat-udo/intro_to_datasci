**Scipy**

Scipy (Scientific python) is a library for scientific computing, with large number of functions to perform various tasks. Most of scipy is based on numpy and numpy array.

```python
# Import various parts of scipy
from scipy import integrate
from scipy import special
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
# calculate integrate x^2 dx from x=0 to x=3
# expected answer is 9
def f1(x):
  return x**2
result = integrate.quad(f1, 0.0, 3.0)
print(result)
```

```python
# calculate integrate 1/(x**2+1) from x=0 to infinity
# expected answer is pi/2
def f2(x):
  return 1.0 / (x**2 + 1)
result = integrate.quad(f2, 0.0, np.inf)
print(result)
```

```python
# Special function
# Get Bessel function of order 2.5 (J_{2.5}(x))
x = np.linspace(0.0, 10.0, 50)
y = special.jv(2.5, x)
plt.plot(x, y, '-ob')
plt.xlabel("x")
plt.ylabel(r"$J_{2.5}(x)$")
plt.show()
```

```python
# Curve fitting
# fit gaussian function to data
def gaussian(x, a, x0, sigma):
  return a * np.exp(-(x - x0)**2 / sigma**2)

xdata = np.linspace(0.0, 5.0, 50)
ydata = 4.5 * np.exp(-(xdata - 2.5)**2 / 1.5**2) + 0.5*np.random.rand(50)
popt, pcov = optimize.curve_fit(gaussian, xdata, ydata)
xplot = np.linspace(0.0, 5.0, 1000)
yplot = gaussian(xplot, popt[0], popt[1], popt[2])
plt.plot(xdata, ydata, 'ob')
plt.plot(xplot, yplot, '-r')
plt.show()
```
